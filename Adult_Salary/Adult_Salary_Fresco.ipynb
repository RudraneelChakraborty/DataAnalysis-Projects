```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Train_df = pd.read_csv('train.csv')
```

```python
Train_df.head()
```

```python
Test_df = pd.read_csv('test.csv')
```

```python
Train_df['age'].max()
```

```python
Train_df.isnull().sum()
```

```python
Train_df = Train_df.dropna()
```

```python
Train_df.drop_duplicates(inplace=True)
```

```python
Train_df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train = Train_df.drop('income_>50K',axis=1)
y_train = Train_df['income_>50K']
X_test = Test_df
```

```python
#Feature Scaling
categorical_columns = ['workclass','education','marital-status','occupation','relationship','race','gender','native-country']
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
age_bins = [0,20,40,60,80,100]
age_labels = [1,2,3,4,5]
X_train['age'] = pd.cut(Train_df['age'], bins=age_bins,labels=age_labels)
X_test['age'] = pd.cut(Test_df['age'], bins=age_bins,labels=age_labels)
```

```python
onehot = OneHotEncoder(sparse=False,dtype='int',drop='first')
encoded_data = onehot.fit_transform(X_train[categorical_columns])
encoded_df = pd.DataFrame(encoded_data,columns = onehot.get_feature_names_out(categorical_columns))
```

```python
X_train = X_train.reset_index(drop=True)
encoded_df = encoded_df.reset_index(drop=True)
```

```python
X_train.drop(columns=categorical_columns,axis=1,inplace=True)
```

```python
X_train = pd.concat([X_train,encoded_df], axis=1)
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
# Feature scaling on X_test
encoded_data_test = onehot.transform(X_test[categorical_columns])
encoded_df_test = pd.DataFrame(encoded_data_test,columns = onehot.get_feature_names_out(categorical_columns))
X_test = X_test.reset_index(drop=True)
encoded_df_test = encoded_df_test.reset_index(drop=True)
X_test.drop(columns=categorical_columns,axis=1,inplace=True)
X_test = pd.concat([X_test,encoded_df_test], axis=1)
```

```python
y_pred = LR.predict(X_test)
```
